#### Data Cleaning

```python
df.columns = df.columns.str.lower()
```

```python
df = df.rename({'pclass':'passenger_class'}, axis = 'columns')
```

```python
del df['ticket']
```

```python
del df['cabin']
```

```python
class_mapping = {1:'first class', 2:'second class', 3:'third class'}
df['passenger_class'] = df['passenger_class'].map(class_mapping)
```

```python
df['last_name'] = df['name'].str.split(',', expand = True)[0]
df['other_name'] = df['name'].str.split(',', expand = True)[1]
```

```python
df
```

```python
df
```

```python
del df['parch']
```

```python
df
```

```python
df['survived'] = df['survived']. replace(0,'Died')
df['survived'] = df['survived']. replace(1,'Alive')
```

```python
df
```

#### BarPlot for Passenger Class

```python
sns.set(style="white")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='passenger_class', hue='sex')
plt.title('Passenger Class Distribution by Gender')
plt.tight_layout()
plt.show()
```

#### PieChart for Gender Distribution

```python
plt.figure(figsize=(6, 6))
gender_counts = df['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])
plt.title('Gender Distribution')
plt.tight_layout()
plt.show()
```

#### Histogram for Age Distribution

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', bins=20, kde=True, color='purple')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Scatter Plot: Fare vs. Age

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='fare', color='orange')
plt.title('Fare vs. Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

#### Box Plot for Fare by Embarked Port

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='embarked', y='fare', palette='Pastel1')
plt.title('Fare Distribution by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

#### Survival by Passenger Class

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='passenger_class', hue='survived', palette='dark')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
